# Real use case of 2D Filtering

```python
from scipy import misc
import matplotlib.pyplot as plt
```

```python
image = misc.face(gray=True)
plt.figure()
plt.imshow(image)
plt.gray()
plt.title('Original')
```

```python
w = np.zeros((50,50))
w[0][0]=10.0
w[49][25] = -10.0
filtered_image = signal.fftconvolve(image,w)
```

```python
plt.figure()
plt.imshow(filtered_image)
plt.gray()
plt.title('Filtered_Image_Conv_2D_example')
```

# 1D filtering 

```python
time = np.arange(0, 10, 0.1)
amplitude = np.sin(time)
plt.plot(time, amplitude)
```

```python
noise = np.random.normal(0,1,100)
```

```python
time.shape
```

```python
noise.shape
```

```python
sine_noise = amplitude+noise
plt.plot(time, sine_noise)
```

```python
sine_noise = amplitude+noise*0.05
plt.plot(time, sine_noise)
```

```python
w = np.zeros(10,np.double)
w[0:5] = 0.09
w[-5:] = 0.9
w
```

```python
filtered_sine = signal.convolve(sine_noise,w,'same')
plt.plot(time, filtered_sine)
```
